# Preprocessing

```python
Q=A.skew()
col_with_skew=list(Q[Q>2].index)
```

```python
col_with_skew
```

```python
import numpy as np
A[col_with_skew].apply(np.log)
```

```python
for j in col_with_skew:
    W=[]
    for i in A[j]:
        if(i != 0):
            W.append(np.log(i))
        else:
            W.append(i)
    A[j]=W
```
